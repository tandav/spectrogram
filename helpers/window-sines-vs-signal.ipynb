```python
import numpy as np
import matplotlib.pyplot as plt
# plt.rcParams['figure.figsize'] = [12, 12]
```

```python
N = 1024
# t_end = 8
t_end = 1
t = np.linspace(0, t_end, N)
# t = np.arange(t_end)
rate = N / t_end
c0 = 0.4 * np.cos(2*np.pi * 4 * t)
s0 = 0.3 * np.sin(2*np.pi * 51 * t)
signal = c0 + s0 

plt.grid()
plt.plot(t, signal, 'k-')
```

```python
f = np.linspace(0, N//4, N*2)
T, F = np.meshgrid(t, f)
E = np.exp(-2j * np.pi * F * T)

corr1 = np.dot(E, signal * np.blackman(N))
corr2 = np.dot(E * np.blackman(N), signal)
corr1 = np.abs(corr1)
corr2 = np.abs(corr2)
```

```python
plt.subplot(1, 3, 1)
plt.plot(f, corr1)
plt.subplot(1, 3, 2)
plt.plot(f, corr2)
plt.subplot(1, 3, 3)
plt.plot(f, corr1 - corr2)
```

```python
np.allclose(corr1, corr2)
```

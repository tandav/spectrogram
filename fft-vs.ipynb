````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
N = 1024
t = np.linspace(0, 1, N)

c0 = np.cos(2*np.pi *   4 * t)
c1 = np.cos(2*np.pi *  10 * t)
c2 = np.cos(2*np.pi * 120 * t)

signal = c0 + c1 + c2
````

````python
f_n = 500
f = np.linspace(0, 200, f_n).reshape(-1, 1)
cosines = np.cos(
    np.ones((f_n, N)) * t * 2 * np.pi * f
)
````

````python
correl = np.sum(cosines * signal, axis=1)
plt.plot(f, correl)
````

````python
difference = np.sum(cosines - signal, axis=1)
plt.plot(f, difference)
````

````python
abs_diff = np.sum(np.abs(cosines - signal), axis=1)
plt.plot(f, abs_diff)
````

````python
summ = np.sum(cosines - signal, axis=1)
plt.plot(f, summ)
````

````python
abs_summ = np.sum(np.abs(cosines + signal), axis=1)
plt.plot(f, abs_summ)
````

````python
diff_sq = np.sum(np.square(cosines - signal), axis=1)
plt.plot(f, diff_sq)
````

````python
summ_sq = np.sum(np.square(cosines + signal), axis=1)
plt.plot(f, summ_sq)
````

````python
prod_sq = np.sum(np.square(cosines * signal), axis=1)
plt.plot(f, prod_sq)
````

````python
harmonic_mean = prod_sq = np.sum(N/(1/cosines + 1/signal), axis=1)
plt.plot(f, harmonic_mean)
````

````python
product_sum = np.product(cosines + signal, axis=1)
plt.plot(f, product_sum)
````

````python
product_prod = np.product(cosines * signal, axis=1)
plt.plot(f, product_prod)
````

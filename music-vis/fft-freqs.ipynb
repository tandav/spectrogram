```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('classic')
import numpy as np
```

```python
rate = 44100
A = 1
f = 50
p = 0
t_end = 1
n = rate * t_end
t = np.linspace(0, 1, n)
s = A * np.sin(2*np.pi*f*t + p)
plt.plot(t, s)
```

```python
frqs = np.fft.rfftfreq(n, d=1/rate)
plt.plot(frqs)
frqs[:1000]
```
